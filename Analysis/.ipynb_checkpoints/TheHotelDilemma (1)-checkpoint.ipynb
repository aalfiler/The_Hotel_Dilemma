```python
# Initial dependencies for data import
import pandas as pd
from pathlib import Path
```

```python
filepath = Path("../UWFinTech_Project3/hotel_bookings.csv")
bookings_df = pd.read_csv(filepath, parse_dates=True, infer_datetime_format=True)
bookings_df.head()
```

```python
# Review data types
bookings_df.dtypes
```

```python
# Identify numbe of rows
bookings_df.count()
```

```python
# Are there null values?
bookings_df.isnull().sum()
```

```python
# Look at percentage nulls for each column
bookings_df.isnull().sum() / len(bookings_df) * 100
```

```python
# Based on the above outcome, we will drop the "Company" column, then fill the nulls in the "Children", "Country", and "deposit_type" columns

# Drop the "company" column
bookings_df.drop(columns=['company'], axis=0, inplace=True)

# The "children" column has a non-material number of nulls & is an interger type, we'll change those to "0"
bookings_df['children'] = bookings_df['children'].fillna("0")

# The "country" columns also has a relatively non-material number of nulls & is an object type, we'll change those to "UNDEF"
bookings_df['agent'] = bookings_df['country'].fillna("UNDEF")

# The "agent" column does have a semi-material number of nulls & is a float type, we'll change those to "1000" which will be unique in that column
bookings_df['country'] = bookings_df['country'].fillna("1000")

# Check updates
bookings_df.isnull().sum()
```

```python
# Change the "object" type columns to numerical data for modelling, we will use "get dummies"

# Question: do I want to do this toreservation_status_date? change to a date format?

model_df = pd.get_dummies(bookings_df, columns=['hotel',
                                               'arrival_date_month',
                                               'meal',
                                               'country',
                                               'market_segment',
                                               'distribution_channel',
                                               'reserved_room_type',
                                               'assigned_room_type',
                                               'deposit_type',
                                               'customer_type',
                                               'reservation_status'])

model_df.describe()
```
